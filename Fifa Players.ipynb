```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('fifa_players.csv')
```

```python
df
```

```python
df.describe()
```

```python
df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
print("Средний возраст:", df['age'].mean())
print("Минимальный возраст:", df['age'].min())
print("Максимальный возраст:", df['age'].max())
```

```python
sns.histplot(df['overall_rating'], bins=20, kde=True)
plt.title('Распределение рейтинга игроков')
plt.xlabel('Overall')
plt.ylabel('Количество игроков')
plt.show()
```

```python
df[df['age'] <= 25][['name', 'age', 'value_euro', 'overall_rating']].sort_values(by='value_euro', ascending=False).head(10).reset_index()
```

```python
df['nationality'].value_counts().head(10)
# 10 самых популярных национальностей
```

```python
# График: зависимость между возрастом и потенциалом
sns.scatterplot(data=df, x='age', y='potential', alpha=0.5)
plt.title('Потенциал в зависимости от возраста')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['value_euro']/1e6, bins=30, kde=True)
plt.xlabel('Стоимость (в миллионах евро)')
plt.ylabel('Количество игроков')
plt.title('Распределение стоимости игроков')
plt.show()
```

```python
#Связь между рейтингом и стоимостью
plt.figure(figsize=(10, 6))
sns.scatterplot(x='overall_rating', y='value_euro', data=df, alpha=0.5)
plt.title('Зависимость стоимости от общего рейтинга')
plt.xlabel('Рейтинг')
plt.ylabel('Стоимость (EUR)')
plt.show()
```

```python
#Самые универсальные игроки — с несколькими позициями
df['num_positions'] = df['positions'].apply(lambda x: len(x.split(',')) if isinstance(x, str) else 0)
df[['name', 'positions', 'num_positions']].sort_values(by='num_positions', ascending=False).head(10).reset_index()
```

```python
top_countries = df['nationality'].value_counts().head(15)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Топ 15 стран по числу игроков')
plt.xlabel('Количество игроков')
plt.ylabel('Страна')
plt.show()
```
